```python
import pandas as pd

adult_income_data = pd.read_csv('adult.csv', na_values='?') # 파일을 보니 none 대신 ? 로 결측치가 표시되어 있음 -> none로 변경
```

```python
adult_income_data.isnull().sum() # 각 열의 결측치 개수 확인
```

```python
adult_income_data.dropna(inplace=True) # 결측치가 있는 행 제거 -- 최대 2800개 밖에 안됨 미미하다고 판단하여 제거
adult_income_data.reset_index(drop=True, inplace=True) # 인덱스 재설정
print(f'결측치 제거 후 행 개수 : {len(adult_income_data)}')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

columns = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
```

```python
for col in columns:
    adult_income_data[col+'_label'] = label_encoder.fit_transform(adult_income_data[col])
```
