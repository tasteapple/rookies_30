```python
import pandas as pd

diamond_data = pd.read_csv('diamonds.csv')
```

```python
diamond_data.info()
```

```python
diamond_data.head()
```

```python
Q1 = diamond_data['carat'].quantile(0.25)
Q3 = diamond_data['carat'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
print(f'상한 : {upper_bound}, 하한 : {lower_bound}')
```

```python
print(f'이상치 제거 전 데이터 갯수 : {len(diamond_data)}')
```

```python
# 이상치 갯수 확인
outliers = diamond_data[(diamond_data['carat'] < lower_bound) | (diamond_data['carat'] > upper_bound)]
print(f'이상치 갯수 : {len(outliers)}')
```

```python
diamond_data = diamond_data[(diamond_data['carat'] >= lower_bound) & (diamond_data['carat'] <= upper_bound)]
print(f'이상치 제거 후 데이터 갯수 : {len(diamond_data)}')
```
